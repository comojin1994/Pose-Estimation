```python
import json
import pandas as pd
import numpy as np
from tqdm import tqdm
from config import *
```

```python
DATA = open(JSON_PATH).readlines()
```

```python
def get_min(datas):
    min_ = 100000
    for data in datas:
        if data == 0: continue
        min_ = min(min_, data)
    return min_

def get_item(data):
    if data['is_train']:
        output = [0] * 37
        output[0] = data['file_name']
        for part in data['parts']:
            output[(int(part) * 2) + 1] = data['parts'][part][0]
            output[(int(part) * 2) + 2] = data['parts'][part][1]
        
        w = max(output[1::2]) - get_min(output[1::2])
        h = max(output[3::2]) - get_min(output[3::2])
        output[33] = data['obj_pos'][0]
        output[34] = data['obj_pos'][1]
        output[35] = w
        output[36] = h
    else:
        output = data['file_name']
        
    return output
```

```python
train = []
test = []

for idx, data in tqdm(enumerate(DATA)):
    data = json.loads(data)
    if data['is_train']:
        if not data['parts']: continue
        train.append(get_item(data))
    else: test.append(get_item(data))
```

```python
train_df = pd.DataFrame(train)
train_df.columns = columns
test_df = pd.DataFrame(test)
test_df.columns = ['image_path']
```
